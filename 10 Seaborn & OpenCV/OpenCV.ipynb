# Face Detection

```python
group_photo = plt.imread('group_photo.jpeg')
```

```python
group_photo.shape
```

```python
plt.imshow(group_photo)
```

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = model.detectMultiScale(group_photo, 1.5)
```

```python
faces.shape
```

```python
faces
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
group_photo = cv2.rectangle(group_photo , (x,y), (x+w, y+h), (255, 0, 0))
```

```python
plt.imshow(group_photo)
```

```python
for one_face in faces:
    x, y, w, h = one_face
    group_photo = cv2.rectangle(group_photo , (x,y), (x+w, y+h), (255, 0, 0), 2)
```

```python
plt.imshow(group_photo)
```
