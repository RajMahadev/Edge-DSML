### Context Managers (with)

```python
with open("example.txt", 'r') as f:
    print("Inside :", f.closed)
    
    
print("Outside :", f.closed)
```

### Writing a file

```python
with open("test.txt", 'w') as f:
    # work with file
    r1 = f.write("This file is generated by Python.\n")
    
    r2 = f.write("Python programming is fun.\n")
    
    print(r1, r2)
```

### Reading a file

```python
with open("test.txt", 'r') as f:
    data = f.read()
    
    print(data)
```

```python
with open("test.txt", 'r') as f:
    data = f.read(5)
    print(data)
    
    next_data = f.read(4)
    print(next_data)
```

#### file pointer
`f.tell()` \
`f.seek(idx)`

```python
with open("test.txt", 'r') as f:
    print(f.tell())
    
    print(f.read(5))
    
    print(f.tell())
    
    print(f.read(4))
    
    
    print(f.tell())
    
    
    # take Fp to 0th index.
    f.seek(0)
    
    print(f.read())
    
```

```python
with open("test.txt", 'r') as f:
    data = f.readline()
    print(data)
```

```python
with open("test.txt", 'r') as f:
    data = f.readlines()
    print(data)
```

```python
with open("test.txt", 'r') as f:
    for line in f:
        print(line)
```

## Binary File

```python
DIC = {"101" : 10000, "105" : 20000, "103": 30000}
```

```python
with open("binary.bin", 'wb') as f:
    f.write(bytes(str(DIC), 'utf-8'))
```

```python
bytes(str(DIC), 'utf-8')
```

```python
with open("binary.bin", 'rb') as f:
    data = f.read()
    print(data)
```
