## Line Plot

$\large y = 2x + 7$


$\large y_2 = (x+5)^2$

```python
x = np.arange(10)
x
```

```python
y = 2*x + 7
y
```

```python
plt.plot(x, y, marker='o')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
y2 = np.square(x + 5)
y2
```

```python
plt.plot(x, y, 'gx--')
plt.plot(x, y2, marker='o')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(y2)
```

```python
plt.scatter(x, y, color='red', marker='*', linewidth=2, label="Nokia")
plt.plot(x, y2, marker='o', label="Apple")
plt.legend()
plt.xlabel("Time (years)")
plt.ylabel("Sales (Million $)")
plt.title("Comparison of Android Vs. Apple")
plt.show()
```

## Bar plot

```python
sales_swift = [13, 19, 25, 11, 15]
sales_nano = [24, 19, 15, 11, 8]
```

```python
sales_nano = np.array(sales_nano)
sales_swift = np.array(sales_swift)
```

```python
x = np.arange(2017, 2022)
x
```

```python
plt.bar(x, sales_swift, width=0.3, label="swift")
plt.bar(x + 0.3, sales_nano, width=0.3, alpha=0.7, label="nano")
plt.legend()
plt.xlabel("years")
plt.ylabel("Sales (Million dollars)")
plt.show()
```

# Pie Chart

```python
marks = np.array([92, 80, 50, 100])
subject = ["Physics", "CS", "Maths", "Science"]
```

```python
faltu = plt.pie(marks, labels=subject, explode = [0,0,0.1,0], shadow=True, autopct='%0.2f')
plt.show()
```

# Histogram

```python
physics_marks = np.round(np.random.randn(50000)*5 + 70)
maths_marks = np.round(np.random.randn(50000)*5 + 80)
```

```python
print(maths_marks[:10])
print(physics_marks[:10])
```

```python
plt.hist(maths_marks, bins=10, label = "Maths")
plt.hist(physics_marks, bins = 10, alpha= 0.5, label="Physics")
plt.legend()
plt.xlabel("Marks")
plt.ylabel("Frequency of students")
plt.savefig('marks.png')
plt.show()
```

# Subplots

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 7
y3 = np.exp(x)
y4 = np.sin(x)
```

```python
plt.subplot(2,2,1)
plt.plot(x, y1)

plt.subplot(2, 2, 2)
plt.plot(x, y2)


plt.subplot(2, 2, 3)
plt.plot(x, y3)

plt.subplot(2, 2, 4)
plt.plot(x, y4)
```
